```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
X, y = make_regression(
    n_samples=100, n_features=1, n_informative=1, noise=15, random_state=99
)
```

```python
X[:2], y[:2]
```

```python
plt.scatter(X, y)
plt.xlabel("Feature")
plt.ylabel("y")
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X, y)

print("Coefficient:", lr.coef_[0])
print("Intercept:", lr.intercept_)
```

```python
class GDRegressor:
    def __init__(self, learning_rate=0.01, epochs=10) -> None:
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.w = 70.20
        self.b = 0

    def fit(self, X, y):
        for i in range(self.epochs):
            slope = -2 * np.sum(y - self.w * X - self.b)
            step_size = self.learning_rate * slope

            self.b = self.b - step_size
```

```python
gr = GDRegressor(learning_rate=0.0001, epochs=1000)
```

```python
gr.fit(X, y)
print("Bias:", gr.b)
```
